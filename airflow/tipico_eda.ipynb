```python
def flatten_dict(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten_dict(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

```python
def get_group(df, column_name, prefix):
    temp_df = df[[column_name, 'id']]
    combined_list = []
    for index, row in temp_df.iterrows():
        nested_dict = row[column_name]
        if isinstance(nested_dict, dict):
            nested_dict_flat = flatten_dict(nested_dict)
            nested_dict_flat['match_id'] = temp_df['id'][index]
            combined_list.append(nested_dict_flat)
    
    group_df = pd.DataFrame(combined_list)
    # Rename columns except 'match_id'
    group_df = group_df.rename(columns={col: f'{prefix}_{col}' if col != 'match_id' else col for col in group_df.columns})
    
    # Ensure no duplicates in the new DataFrame using both 'group_id' (renamed) and 'match_id'
    group_df = group_df.drop_duplicates(subset=[f'{prefix}_id', 'match_id'])
#group_df = get_group(df, 'group', 'group')

# Drop the original nested column from the original DataFrame
    df = df.drop(columns=['group'])

    # Merge the extracted DataFrame back into the original DataFrame
    merged_df = df.merge(group_df, left_on='id', right_on='match_id').drop(columns=['match_id'])

    
    return merged_df
```
