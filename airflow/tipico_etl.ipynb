```python
import requests
#from airflow.utils.dates import days_ago
import requests
import csv
from datetime import datetime
import pandas as pd
import os
```

```python
url = 'https://trading-api.tipico.us/v1/pds/lbc/events/live?licenseId=US-NJ&lang=en&limit=18'
response = requests.get(url)
data = response.json()
```

```python
df = pd.json_normalize(data[0])
```

```python
df
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df["participants"]
```

```python
participants_df = df['participants'].apply(pd.Series)
```

```python
participants_eachone_df = participants_df[0].apply(pd.Series)
participants_eachone_df['match_id'] = df['id']
```

```python
participants_eachone_df
```

```python
participant_temp_df = participants_df[1].apply(pd.Series)
participant_temp_df['match_id'] = df['id']
```

```python
participants_df
```

```python
participants_df = pd.concat([participants_eachone_df, participant_temp_df], ignore_index=True)
```
